```python
import locale
import openai
import yaml
import os
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage
import panel as pn
from IPython.display import display, Markdown,Latex,Image

# Das sind spezielle Formatierungsvorlagen
pn.extension('katex',design="material", sizing_mode="stretch_width")
locale.getpreferredencoding = lambda: "UTF-8"

# Hier werden die API Schlüssel eingelesen
with open('chatgpt_api_credentials.yml', 'r') as file:
    api_creds = yaml.safe_load(file)

# Hier wähle ich das LLLM aus, welche ich weiter unten benutze
api_creds.keys()
os.environ['OPENAI_API_KEY'] = api_creds['openai_key']
openai.api_key  = api_creds['openai_key']
model= ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
#model = ChatOpenAI(model='gpt-4o',temperature=0)
```

# Use Prompt Templates and run LLM.

```python

SYS_PROMPT = f"""You are an expert in Mathematics, specializing in creating high-quality,
real-world problems for high school students. Your task is to generate challenging and engaging problems that help students apply 
mathematical concepts to practical situations.
            """

USER_PROMPT =f"""
             Generate a problem involving matrices and determinants.
             Choose an application in economics.
            """

messages = [
SystemMessage(content=SYS_PROMPT), 
HumanMessage(content=USER_PROMPT) 
]
response = model.invoke(messages)       
```

```python
# Hier spukt es die ganzen Daten un einem Json Datenformat aus
print(response)
```

```python
# Ich bin nur an dem Inhalt der Antwort, nicht dem ganzen drumherum interessiert
print(response.content)
```
